```python
# It will print 0-9
for i in range(10):
  print(i)
```

```python
# It will print 1-9
for i in range(1, 10):
  print(i)
```

```python
# It will print odd numbers between 1-9
for i in range( 1, 10, 2):
  print(i)
```

```python
# It will print even numbers between 10 to 2 in reverse way
for i in range( 10, 1, -2):
  print(i)
```
