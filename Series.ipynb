```python
1^2 + 2^2 + 3^2 + ....x^2

x=input("Enter Last Number: ")

sum=0
for i in range(1, int(x)+1):
  sum=sum+(i*i)
print(sum)
```

```python
#1^3 + 2^3 + 3^3 + ....x^3

x=input("Enter Last Number: ")

sum=0
for i in range(1, int(x)+1):
  sum=sum+(i*i*i)

print(f"Sum of 1^3 + 2^3 + ....{x}^3 = ", sum)

```

```python
#2^3 + 4^3 + 6^3 + ....x^3

x=input("Enter Last Number: ")

sum=0
for i in range(2,int(x)+1 ,2):
  sum=sum+(i*i*i)

print(f"Sum of 2^3 + 4^3 + ....{x}^3 = ", sum)
```

```python


#1/2 + 1/2 + 1/3 + ....1/x

x=input("Enter Last Number: ")

sum=0
for i in range(1, int(x)+1):
  sum=sum+(1/i)
print(sum)
```

```python
#1/x + 2/x + 3/x + ....x^x

x=input("Enter Last Number: ")

sum=0
for i in range(1, int(x)+1):
  sum=sum+(i/int(x))            # ex: 1/5 + 2/5 + 3/5 + 4/5 + 5/5 = 3
print(sum)
```

```python
# 1*2*3....*x

x=input("Enter Last Number: ")

mul=1
for i in range(1, int(x)+1):
  mul=mul*i
print(f"1*2*3....*{x} = ",mul)
```
