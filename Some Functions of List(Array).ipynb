```python
#lenght of list

fruits=["Apple","Mango","Jackfruit", "Banana","Berry","Guava","Orange"]
print("Lenght: ", len(fruits))


```

```python
# Add New Item in the list (at last index)

fruits=["Apple","Mango","Jackfruit", "Banana","Berry","Guava","Orange"]

fruits.append("Blueberry")
print(fruits)

```

```python
# Add New Item in the list (at user defined index)

fruits=["Apple","Mango","Jackfruit", "Banana","Berry","Guava","Orange"]

fruits.insert(2,"Blueberry")
print(fruits)

```

```python
fruits=["Apple","Mango","Jackfruit", "Banana","Berry","Guava","Orange"]

fruits.remove("Jackfruit")
print(fruits)

```

```python
#Sort all items in ascending order
fruits=["Apple","Mango","Jackfruit", "Banana","Berry","Guava","Orange"]

fruits.sort()
print(fruits)

num=[10,82,92,8,29,53,29,49,5,18]
num.sort()
print(num)

```

```python
#Clear last item from list

fruits=["Apple","Mango","Jackfruit", "Banana","Berry","Guava","Orange"]

fruits.pop()
print(fruits)


```

```python
#Clear all item from list

fruits=["Apple","Mango","Jackfruit", "Banana","Berry","Guava","Orange"]

fruits.clear()
print(fruits)

```

```python
#Copy all item from list

fruits=["Apple","Mango","Jackfruit", "Banana","Berry","Guava","Orange"]

fr = fruits.copy()
print(fr)

```

```python
#Find out index number of a item from list

fruits=["Apple","Mango","Jackfruit", "Banana","Berry","Guava","Orange"]

pos = fruits.index("Banana")
print(pos)

```

```python
#how many times 1 item exists in the list

fruits=["Banana","Mango","Jackfruit", "Banana","Berry","Guava","Banana"]

print(fruits.count("Banana"))
print(fruits.count("Mango"))
```
