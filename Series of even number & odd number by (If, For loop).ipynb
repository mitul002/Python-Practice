```python
#Even Number
last_no = input("Enter the last number: " )

sum= 0

for i in range(int(last_no)+1):
  if i%2==0:   # by which value of i, the condition is true that i will be taken. ex: 2,4,6...
    sum=sum+i
print(sum)
```

```python
#Odd Number
last_no = input("Enter the last number: " )

sum= 0

for i in range(int(last_no)+1):
  if (i%2!=0):      # by which value of i, the condition is true that i will be taken. ex: 1,3,5...
    sum=sum+i
print(sum)
```
