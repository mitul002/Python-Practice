```python
def student(**Address):    #xxargs works with dictionary
  print(Address)

student(Name="Jack", ID=1232, Phone= 1526627282)
student(Name="John", ID=4267, Phone= 1573903030)
student(Name="July", ID=2425, Phone= 1277388229)
```
