```python
num1 = input("Enter an Integer number: ")
num2 = float(input("Enter a Float number: "))

sum= int(num1)+ float(num2)
print("Sum: ", sum)
print("Sub: ", int(num1)-(num2))
print("Mul: ", int(num1)*(num2))
print("sub: ", int(num1)/(num2))
print("Mod: ", int(num1)%(num2))
print("Exponential: ", int(num1)**(num2))
```
