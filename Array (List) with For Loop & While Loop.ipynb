```python
#Using While loop

fruits=["Apple","Banana", "Guava", "Berry", "Dragonfruit"]

i=0
while i<=4:
  print(fruits[i])
  i=i+1
```

```python
#Using While loop with len function

fruits=["Apple","Banana", "Guava", "Berry", "Dragonfruit"]
length=len(fruits)

i=0
while i<length:
  print(fruits[i])
  i=i+1

```

```python
fruits=["Apple","Banana", "Guava", "Berry", "Dragonfruit"]

for i in fruits:
  print(i)
```
