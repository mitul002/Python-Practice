```python
def student(name, id, phone):
  print(name,id,phone)

student("Jack", 1232, 1526627282)
student("John", 4267, 1573903030)
student("July", 2425, 1277388229)

```

```python
def student(*Address):    #xargs works with tuples

  print(Address)

student("Jack", 1232, 1526627282)
student("John", 4267, 1573903030)
student("July", 2425, 1277388229)
```

```python

def sum(*number):

  sum=0
  for i in number:
    sum=sum+i
  print(sum)

sum(10,35,37)
sum(52,73,737,37,27,27)
```
