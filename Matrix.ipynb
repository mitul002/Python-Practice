```python
matrix = [
          [1, 2, 3],
          [4, 6, 5],
          [6, 7, 9]
         ]

print(matrix)

print(matrix[0][0])
print(matrix[0][2])

#Change matrix value
matrix[2][2]=8    #previous value 9
print(matrix[2][2])
```
