```python
#Stack works in "first in first out method"

from collections import deque
list1=deque(['Book 5', 'Book 6', 'Book 7', 'Book 8'])
print(list1)


list1.popleft()
print(list1)

list1.popleft()
print(list1)

list1.popleft()
list1.popleft()
if not list1:     #If no book found in the list then this message will print
  print("Zero book left")
```
