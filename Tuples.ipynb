```python
"""
List and Tuples are almost same.
But we can't change value of tuples
"""

'''
We use [] in list(array)
we use {} in dictionary
we use () in tuples
'''

tup=(

     "Hey Google",
     "Hi Google",
     "Ok Google",
     ("Bye Google", 134, 56)
)

print(tup[1])
print(tup[2])

print(tup[3][1])   #row 3, column 1

```
