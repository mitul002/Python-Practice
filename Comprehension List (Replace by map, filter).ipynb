```python
#using map function

num=[2,3,5,7]
result=list(map(lambda a: a*a, num))
print(result)
```

```python
#Using Comprehension List Function

num=[2,3,5,7]

result= [a*a for a in num] #[expression for item in list]
print(result)
```

```python
#Using Comprehension List Function

result= [x*x*x for x in num] #[expression for item in list]
print(result)


num=[2,5,8,9,12,15,20,22]
```

```python




#using filter function
num = [1,2,3,4,5,6,7,8,9]

result= list(filter(lambda x: x%2==0, num))
print(result)
```

```python
#Using Comprehension List Function
num = [1,2,3,4,5,6,7,8,9]

result=[x for x in num if x%2==0]

print(result)
```
